# Spliting the data into Independent and Dependent Variable

```python
X = df.drop(columns = 'Outcome', axis = 1)
y = df['Outcome']
```

```python
X.columns
```

```python
X.shape
```

```python
y.shape
```

# Data Preprocessing

Standardization of data :

It is used to do because we can see that data have different range in the different column so it will be difficult for are ml model to predict so will we standardize it.

So it will range all the data into 0 and 1

```python
scalar = StandardScaler()
scalar.fit(X)
std_data = scalar.transform(X)
```

```python
print(std_data)
```

# Creating Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, test_size = 0.2, random_state=42)
```

# Model Creation

```python
classifier = svm.SVC(kernel='linear')
classifier.fit(X_train, y_train)
```

# Model Evaluation

Accuracy Score

```python
train_acc = classifier.predict(X_train) 
train_data_acc = accuracy_score(train_acc, y_train)
print('Training Data Accuracy : ', train_data_acc)
```

```python
test_acc = classifier.predict(X_test) 
test_data_acc = accuracy_score(test_acc, y_test)
print('Testing Data Accuracy : ', test_data_acc)
```
