# Import libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

# Importing Dataset 

```python
df = pd.read_csv('dataset\parkinsons.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# distribution of the people have parkisons diseases
df['status'].value_counts()
```

```python
df['status'].value_counts().mean()
```

1 -> Person have Parkison's

0 -> Person do not have Parkison's

```python
df.groupby('status').mean()
```

```python
# Spliting the data into independent and dependent variable, we are droping column with axis 1 and row axis 0

X = df.drop(columns = ['status', 'name'], axis = 1)
y = df['status']
```

```python
print(y)
```

```python
X.shape
```

```python
y.shape
```

# Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, test_size = 0.2, random_state=42)
```

# Data Standardization

This is done when we have multiple data in different values.

```python
scalar = StandardScaler()
```

```python
X_train = scalar.fit_transform(X_train)
```

```python
X_train
```

# Model Creation

SVM -> Support Vector Machine

```python
model = svm.SVC(kernel = 'linear')
model.fit(X_train, y_train)
```

```python
# accuracy acore in train data
X_pred = model.predict(X_train)
X_acc = accuracy_score(X_pred, y_train)
```

```python
print("Accuracy on Train Data : ", X_acc)
```

```python
# accuracy acore in test data
X_pred = model.predict(X_test)
X_acc = accuracy_score(X_pred, y_test)
```

```python
print("Accuracy on Test Data : ", X_acc)
```

# Creating Prediction

```python
input_data = (116.67600,	137.87100,	111.36600,	0.00997,	0.00009,	0.00502,	0.00698,	0.01505,	0.05492,	0.51700,	0.02924,	0.04005,	0.03772,	0.08771,	0.01353,	20.64400,	0.434969,	0.819235,	-4.117501,	0.334147,	2.405554,	0.368975)

# changing the input_data into numpy array
inp_arr = np.asarray(input_data)

# reshaping the inp_arr as we are predicting for only one instance
inp_reshape = inp_arr.reshape(1, -1)

prediction = model.predict(inp_reshape)
print("Prediction of the Parkisons Diseases is : ", prediction)

if (prediction == [0]):
  print('Person do not have Parkisons Diseases...')
else :
  print('Person have Parkisons Diseases...')
```
